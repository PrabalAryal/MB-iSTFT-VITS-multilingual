```python
import torch, librosa
p = "nl_audio/09362.wav"
print("spec path:", p.replace(".wav", ".spec.pt"))
# check spec file if present
spec_path = p.replace(".wav", ".spec.pt")
if os.path.exists(spec_path):
    spec = torch.load(spec_path)
    print("spec.shape:", getattr(spec, "shape", spec))
else:
    y, sr = librosa.load(p, sr=22050, mono=True)
    import numpy as np
    import librosa
    D = np.abs(librosa.stft(y, n_fft=1024, hop_length=256, win_length=1024))
    print("recomputed stft shape:", D.shape, "wav len:", len(y), "sr:", sr)
```

```python
import os, torch
DATA_DIR = "nl_audio"
bad = []
for root, _, files in os.walk(DATA_DIR):
    for f in files:
        if not f.lower().endswith(".spec.pt"): continue
        p = os.path.join(root, f)
        try:
            s = torch.load(p)
            if (not hasattr(s,'dim')) or s.dim()==1:
                bad.append((p, '1D'))
            elif s.size(1) <32:   # 32 frames = segment_size // hop_length from your config
                bad.append((p, f'frames={s.size(1)}'))
        except Exception as e:
            bad.append((p, f'load_err:{e}'))
if bad:
    print("PROBLEMS:")
    for b in bad: print(b)
else:
    print("All specs OK")
```
